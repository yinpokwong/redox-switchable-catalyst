```python
xyz_match = ['X           Y           Z']
LUMO_match = ['Alpha virt. eigenvalues']
HOMO_match = ['Alpha  occ. eigenvalues']
nbo_match = ['Natural Population Analysis']
mulliken_match = ['Mulliken charges:']

missing_nbo_spe_calc =[]
missing_mulliken_spe_calc = []

# Initiate list for gathering descriptor values
name_list = []
metal_atom_list = []
donor_atom_1_list = []
donor_atom_2_list = []

metal_d1_distance_list = []
metal_d2_distance_list = []
metal_o1_distance_list = []
metal_o2_distance_list = []
metal_ome_distance_list = []
metal_fe_distance_list = []

lumo_list = []
homo_list = []
homo_lumo_list = []

metal_nbo_list = []
d1_nbo_list = []
d2_nbo_list = []
o1_nbo_list = []
o2_nbo_list = []
ome_nbo_list = []
fe_nbo_list = []

nbo_d1_d2_avg_list = []
nbo_o1_o2_avg_list = []
nbo_fe_m_diff_list = []


```

```python
# Read in .out file from spe file
index_count = 0
for subdir,dirs,files in os.walk(spe_directory):                  # Loop over each directory, subdirectory and files
    for file in files:                                      # Loop over each file
        if any([file.endswith('-spe.out')]):                    # If file is a .out file
            filename = os.path.join(subdir, file)       # Return path to file
            name = Path(filename).stem           # Extract filename from the end of path and return as a string
            name = name.replace('-spe','')     #Remove -opt in string to get just the name of ligand 'pd-x-x-x'
            name_list.append(name)

            mylines = []

            with open (filename, 'rt') as myfile:       # Open .out for reading text
                # myfile = myfile.read()                # Read the entire file to a string
                for myline in myfile:                    # For each line, stored as myline,
                    mylines.append(myline)               # add its contents to mylines list.

                    
                for line in mylines:
                    if 'NAtoms' in line:
                        number_list = re.findall('-?\d*\.?\d+',line)            # get number of atoms
                        natoms = int(number_list[0])   
                        print(name, natoms)
                        break



                # Identify row in atom_label_df that matches the name
                matching_row = atoms_label_df.loc[atoms_label_df['Filename'] == name]
                # print(matching_row)


                # Assign atom labels based on dataframe 
                # print(index_count)
                metal_atom = matching_row.iloc[0]['m']
                donor_atom_1 = matching_row.iloc[0]['d1']
                donor_atom_2 = matching_row.iloc[0]['d2']
                oxygen_1 = matching_row.iloc[0]['o1']
                oxygen_2 = matching_row.iloc[0]['o2']
                ome_atom = matching_row.iloc[0]['ome']
                fe_atom = matching_row.iloc[0]['fe']


                for line in mylines:
                    for phrase in xyz_match:                                # iterate through each phrases
                        if phrase in line:                                          # check if phrase is in line

                                                                                    # For loop for generating the XYZ coordinates
                            count = 0
                            xyz = []
                            line_number = mylines.index(line) + 2
                            while count < natoms:
                                count = count + 1
                                xyz.append(mylines[line_number])
                                line_number = line_number + 1


                x_coord = []
                y_coord = []
                z_coord = []
                atom_symbol =[]
                element = { "1": "H", "2": "He", "3": "Li", "4": "Be", "5": "B", "6": "C", "7": "N", "8": "O", "9": "F", "10": "Ne",
                            "11": "Na", "12": "Mg", "13": "Al", "14": "Si", "15": "P", "16": "S", "17": "Cl", "18": "Ar", "19": "K", "20": "Ca",
                            "21": "Sc", "22": "Ti", "23": "V", "24": "Cr", "25": "Mn", "26": "Fe", "27": "Co", "28": "Ni", "29": "Cu", "30": "Zn",
                            "31": "Ga", "32": "Ge", "33": "As", "34": "Se", "35": "Br", "36": "Kr", "37": "Rb", "38": "Sr", "39": "Y", "40": "Zr",
                            "41": "Nb", "42": "Mo", "43": "Tc", "44": "Ru", "45": "Rh", "46": "Pd", "47": "Ag", "48": "Cd", "49": "In", "50": "Sn",
                            "51": "Sb", "52": "Te", "53": "I", "54": "Xe", "55": "Cs", "56": "Ba", "57": "La", "58": "Ce", "59": "Pr", "60": "Nd",
                            "61": "Pm", "62": "Sm", "63": "Eu", "64": "Gd", "65": "Tb", "66": "Dy", "67": "Ho", "68": "Er", "69": "Tm", "70": "Yb",
                            "71": "Lu", "72": "Hf", "73": "Ta", "74": "W", "75": "Re", "76": "Os", "77": "Ir", "78": "Pt", "79": "Au", "80": "Hg",
                            "81": "Tl", "82": "Pb", "83": "Bi", "84": "Po", "85": "At", "86": "Rn", "87": "Fr", "88": "Ra", "89": "Ac", "90": "Th",
                            "91": "Pa", "92": "U", "93": "Np", "94": "Pu", "95": "Am", "96": "Cm", "97": "Bk", "98": "Cf", "99": "Es", "100": "Fm",
                            "101": "Md", "102": "No", "103": "Lr", "104": "Rf", "105": "Db", "106": "Sg", "107": "Bh", "108": "Hs", "109": "Mt",
                            "110": "Ds", "111": "Rg", "112": "Cn", "113": "Nh", "114": "Fl", "115": "Mc", "116": "Lv", "117": "Ts", "118": "Og"}

                # Generate XYZ file in .txt form, then find xyz coordinates for metal, atom_a and atom_b
                for line in xyz:
                    number_list = re.findall('-?\d*\.?\d+',line)
                    atom_number = int(number_list[0])
                    element_number = int(number_list[1])
                    atom_x = float(number_list[3])
                    atom_y = float(number_list[4])
                    atom_z = float(number_list[5])

                    # Make xyz coord into .txt file for percent buried volume
                    x_coord.append(atom_x)
                    y_coord.append(atom_y)
                    z_coord.append(atom_z)
                    atom_symbol.append(element[str(element_number)])
                    name_xyz = name + '_xyz.txt'            

                    # Identify metal, d1, d2, o1, o2, ome and fe sequence numbers and obtain their respective x,y,z coords
                    if atom_number == metal_atom:
                        metal_x = atom_x
                        metal_y = atom_y
                        metal_z = atom_z
                    elif atom_number == donor_atom_1:
                        d1_x = atom_x
                        d1_y = atom_y
                        d1_z = atom_z
                    elif atom_number == donor_atom_2:
                        d2_x = atom_x
                        d2_y = atom_y
                        d2_z = atom_z
                    elif atom_number == oxygen_1:
                        o1_x = atom_x
                        o1_y = atom_y
                        o1_z = atom_z
                    elif atom_number == oxygen_2:
                        o2_x = atom_x
                        o2_y = atom_y
                        o2_z = atom_z
                    elif atom_number == ome_atom:
                        ome_x = atom_x
                        ome_y = atom_y
                        ome_z = atom_z                  
                    elif atom_number == fe_atom:
                        fe_x = atom_x
                        fe_y = atom_y
                        fe_z = atom_z              

                data = {
                    'Atom': atom_symbol,
                    'X': x_coord,
                    'Y': y_coord,
                    'Z': z_coord
                }

                xyz_df = pd.DataFrame(data)      
                # xyz_df.to_csv(name_xyz, header=False, index=False, sep = " ")

                # Find HOMO - LUMO energy gap in SPE file

                LUMO = []
                for phrase in LUMO_match:
                    for line in mylines:
                        if phrase in line:
                            LUMO.append(line)
                number_list = re.findall('-?\d*\.?\d+',LUMO[0])            # get value from line containing LUMO
                LUMO_energy = float(number_list[0])                         # get first value from the line, which represents the LUMO

                HOMO = []
                for phrase in HOMO_match:
                    for line in mylines:
                        if phrase in line:
                            HOMO.append(line)
                number_list = re.findall('-?\d*\.?\d+',HOMO[-1])            # get value from line containing LUMO
                HOMO_energy = float(number_list[-1])                        # get lsat value from line, which represents HOMO

                HOMO_LUMO = round(LUMO_energy - HOMO_energy,4)    
                print('LUMO: ', LUMO_energy)
                print('HOMO: ', HOMO_energy)
                print('HOMO_LUMO: ', HOMO_LUMO)


                # Find NBO charge for metal, d1, d2, o1, o2, ome, fe
                # Obtain indexes for calculating NBO charge of different atom distances

                for phrase in nbo_match:                                       # iterate through each phrases
                    nbo_match_count = 0
                    for line in mylines:
                        if phrase in line:                                          # check if phrase is in line
                            nbo_match_count = nbo_match_count + 1
                            count = 0
                            nbo_xyz = []
                            line_number = mylines.index(line) + 6
                            while count < natoms:
                                count = count + 1
                                nbo_xyz.append(mylines[line_number])
                                line_number = line_number + 1

                            with open('nbo.txt', 'w') as filehandle:                   # Save nbo_xyz  into a txt file
                                for listitem in nbo_xyz:
                                    filehandle.write(listitem)

                    # Extract NBO charge for metal, d1, d2, o1, o2, ome, fe

                    for line in nbo_xyz:
                        number_list = re.findall('-?\d*\.?\d+',line)
                        print(number_list)
                        atom_number = int(number_list[0])

                        if atom_number == metal_atom:
                            metal_nbo = float(number_list[1])
                            print('metal_nbo: ', metal_nbo)

                        elif atom_number == donor_atom_1:
                            d1_nbo = float(number_list[1])
                            print('donor atom 1 nbo: ', d1_nbo)

                        elif atom_number == donor_atom_2:
                            d2_nbo = float(number_list[1]) 
                            print('donor atom 2 nbo: ', d2_nbo)
                        
                        elif atom_number == oxygen_1:
                            o1_nbo = float(number_list[1]) 
                            print('oxygen 1 nbo: ', o1_nbo)
                        
                        elif atom_number == oxygen_2:
                            o2_nbo = float(number_list[1]) 
                            print('oxygen 2 nbo: ', o2_nbo)

                        elif atom_number == ome_atom:
                            ome_nbo = float(number_list[1]) 
                            print('ome nbo: ', ome_nbo)
                        
                        elif atom_number == fe_atom:
                            fe_nbo = float(number_list[1]) 
                            print('fe nbo: ', fe_nbo)

                    nbo_d1_d2_avg = round((d1_nbo+d2_nbo)/ 2, 4)
                    nbo_o1_o2_avg = round((o1_nbo+o2_nbo)/ 2, 4)
                    nbo_fe_m_diff = round(abs(fe_nbo - metal_nbo), 4)
                        
    #           Extract metal - donor atom distance

                metal_d1_distance = round(math.sqrt((metal_x - d1_x) ** 2 + (metal_y - d1_y) ** 2 + (metal_z - d1_z) ** 2),3)    
                metal_d2_distance = round(math.sqrt((metal_x - d2_x) ** 2 + (metal_y - d2_y) ** 2 + (metal_z - d2_z) ** 2),3)
                metal_o1_distance = round(math.sqrt((metal_x - o1_x) ** 2 + (metal_y - o1_y) ** 2 + (metal_z - o1_z) ** 2),3)    
                metal_o2_distance = round(math.sqrt((metal_x - o2_x) ** 2 + (metal_y - o2_y) ** 2 + (metal_z - o2_z) ** 2),3)
                metal_ome_distance = round(math.sqrt((metal_x - ome_x) ** 2 + (metal_y - ome_y) ** 2 + (metal_z - ome_z) ** 2),3)
                metal_fe_distance = round(math.sqrt((metal_x - fe_x) ** 2 + (metal_y - fe_y) ** 2 + (metal_z - fe_z) ** 2),3)


                # Append each descriptor into the list 
                metal_atom_list.append(metal_atom)
                donor_atom_1_list.append(donor_atom_1)
                donor_atom_2_list.append(donor_atom_2)
                metal_d1_distance_list.append(metal_d1_distance)
                metal_d2_distance_list.append(metal_d2_distance)
                metal_o1_distance_list.append(metal_o1_distance)
                metal_o2_distance_list.append(metal_o2_distance)
                metal_ome_distance_list.append(metal_ome_distance)
                metal_fe_distance_list.append(metal_fe_distance)
                lumo_list.append(LUMO_energy)
                homo_list.append(HOMO_energy)
                homo_lumo_list.append(HOMO_LUMO)
                metal_nbo_list.append(metal_nbo)
                d1_nbo_list.append(d1_nbo)
                d2_nbo_list.append(d2_nbo)
                o1_nbo_list.append(o1_nbo)
                o2_nbo_list.append(o2_nbo)
                ome_nbo_list.append(ome_nbo)
                fe_nbo_list.append(fe_nbo)
                nbo_d1_d2_avg_list.append(nbo_d1_d2_avg)
                nbo_o1_o2_avg_list.append(nbo_o1_o2_avg)
                nbo_fe_m_diff_list.append(nbo_fe_m_diff)

# Define data dictionary for excel spreadsheet

data = {'Name': name_list,
        'Metal-atom' : metal_atom_list,
        'Donor-atom-1' : donor_atom_1_list,
        'Donor-atom-2' : donor_atom_2_list,
        'Metal-d1-distance' : metal_d1_distance_list,
        'Metal-d2-distance' : metal_d2_distance_list,
        'Metal-o1-distance' : metal_o1_distance_list,        
        'Metal-o2-distance' : metal_o2_distance_list,
        'Metal_ome_distance' : metal_ome_distance_list,
        'Metal_fe_distance' : metal_fe_distance_list,
        'LUMO' : lumo_list,
        'HOMO' : homo_list,
        'HOMO_LUMO' : homo_lumo_list,
        'Metal_nbo' : metal_nbo_list,
        'D1_nbo' : d1_nbo_list,
        'D2_nbo' : d2_nbo_list,
        'O1_nbo' : o1_nbo_list,
        'O2_nbo' : o2_nbo_list,
        'OMe_nbo' : ome_nbo_list,
        'Fe_nbo' : fe_nbo_list,
        'D1_d2_nbo_avg' : nbo_d1_d2_avg_list,
        'O1_o2_nbo_avg' : nbo_d1_d2_avg_list,   
        'Fe_m_diff_avg' : nbo_fe_m_diff_list,
       }
# Check the lengths of each list in the dictionary
for key, value in data.items():
    print(f"Length of {key}: {len(value)}")

descriptor_data = pd.DataFrame(data)
csv_filename = 'descriptor_data.csv'
descriptor_data.to_csv(csv_filename,index=False) 
```

```python
# Put back virtual label in descripor_data dataframe
# Assuming descriptor_data and atoms_label_df are already defined

# Create a dictionary to map Filename to virtual value
filename_to_virtual = dict(zip(atoms_label_df['Filename'], atoms_label_df['virtual']))

# Define a function to map Name to virtual value based on filename_to_virtual dictionary
def map_virtual(name):
    return filename_to_virtual.get(name, 'no')  # Default to 'no' if no match is found

# Apply the function to the 'Name' column in descriptor_data and create the 'virtual' column
descriptor_data['virtual'] = descriptor_data['Name'].apply(map_virtual)

# Display the updated DataFrame
descriptor_data
```

```python
# Split into two dataset, depending on whether it's red/ox
# Split the dataframe based on the suffix in the 'Name' column
descriptor_data_ox = descriptor_data[descriptor_data['Name'].str.endswith('-ox')]
descriptor_data_red = descriptor_data[descriptor_data['Name'].str.endswith('red')]

# # Remove the '-ox' suffix from the 'Name' column in descriptor_data_ox
descriptor_data_ox['Name'] = descriptor_data_ox['Name'].str.rstrip('-ox')

# # Remove the '-red' suffix from the 'Name' column in descriptor_data_red
descriptor_data_red['Name'] = descriptor_data_red['Name'].str.replace('-red$', '', regex=True)

descriptor_data_red
```

```python
column_list = descriptor_data.columns.tolist()
# Step 2: Remove 'Name' and 'virtual' from the list of columns
columns_to_keep = [col for col in column_list if col not in ['Name', 'virtual','Metal-atom','Donor-atom-1','Donor-atom-2']]
columns_to_keep
```

```python
# Assuming descriptor_data_ox and descriptor_data_red are your DataFrames

# Step 1: Merge the two DataFrames on the 'Name' column
merged_data = pd.merge(descriptor_data_ox, descriptor_data_red, on='Name', suffixes=('_ox', '_red'))

# Step 2: Calculate the difference between corresponding descriptor columns
# List all columns that should have differences calculated
descriptor_columns = columns_to_keep
# Initialize an empty dictionary to store the differences
difference_data = {'Name': merged_data['Name']}

# Compute differences for each column
for col in descriptor_columns:
    difference_data[col] = merged_data[f'{col}_ox'] - merged_data[f'{col}_red']

# Step 3: Create a new DataFrame to store the differences
descriptor_difference_df = pd.DataFrame(difference_data)

# Display the result
descriptor_difference_df
```
